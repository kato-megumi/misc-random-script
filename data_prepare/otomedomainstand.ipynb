```python
from os.path import join
from tqdm import tqdm
import random
from PIL import Image
import csv
import os
from pathlib import Path

directory = "/Users/khoi.ho/Downloads/Data/"
chara = ["ひなた", "七海", "柚子", "湊", "風莉", "美結"]
pose = ["Ａ", "Ｂ", "Ｃ"]
level = ["_0_", "_1_", "_2_", "_3_"]

for l in level:
    if not os.path.exists(os.path.join(directory, l)):
        os.mkdir(os.path.join(directory, l))

sprite_dict = {key: {p: {"stand": [], "other": []} for p in pose} for key in chara[1:]}
sprite_dict["ひなた"] = {p: {"stand": [], "other": []} for p in ["Ａ２", "Ｂ２", "Ｃ"]}

for c, c_dict in sprite_dict.items():
    for p in c_dict.keys():
        dir = join(directory, c)
        for filename in os.listdir(dir):
            filepath = os.path.join(dir, filename)
            if os.path.isfile(filepath) and Path(filename).name.startswith(c+p+level[0]) and os.path.getsize(filepath) > 1048576:
                id = os.path.splitext(filename)[0].split("_")[-1]
                sprite_dict[c][p]["stand"].append(id)

sprite_dict["ひなた"]["Ａ２"]["other"] = [['2694', '2590', '2569', '2464', '2440', '2357', '2273', '2242', '2161'],['10152', '10153', '10154', '10155', '10156'],['10177', '10182', '10187', '10192', '10197', '10202', '10207', '10212', '10219']]
sprite_dict["ひなた"]["Ｂ２"]["other"] = [['8184', '8197', '8210', '8220', '8231', '8240', '8248', '8258'],['6498', '6532', '6564', '6582', '6616', '6636', '6656', '6691', '6726'],['9314', '9319', '9324', '9329', '9334', '9339', '9344', '9349', '9373', '9380', '9390']]
sprite_dict["柚子"]["Ａ"]["other"] = [['3501', '3536', '3559', '3594', '3631', '3661'],['4821', '4822', '4823', '4824', '4825', '4826', '4827', '4828'],['4855', '4860', '4865', '4885', '4880', '4875', '4870', '4890', '4897', '4902', '4907']]
sprite_dict["柚子"]["Ｂ"]["other"] = [['1984', '2029', '2051', '2098', '2135', '2173'],['5557', '5558', '5559', '5561', '5562', '5563', '5564', '5565', '5567', '5573'],['5599', '5594', '5589', '5584', '5579', '5573', '5604', '5609', '5616', '5621', '5626', '5631']]
sprite_dict["湊"]["Ａ"]["other"] = [['2172', '2224', '2270', '2316', '2353', '2386', '2409', '2437', '2460'],['3825', '3826', '3827', '3828', '3829', '3830', '3831'],['3851', '3856', '3861', '3866', '3871', '3876', '3881', '3888', '3893', '3898', '3903', '3908']]
sprite_dict["湊"]["Ｂ"]["other"] = [['1775', '1800', '1848', '1899', '1935', '1983', '2007', '2043'],['4177', '4162', '4163', '4164', '4165', '4166', '4167', '4168', '4159', '4160', '4161'],['4180', '4183', '4188', '4193', '4198', '4203', '4208', '4213', '4218', '4225', '4230']]
sprite_dict["風莉"]["Ａ"]["other"] = [['5746', '5755', '5766', '5777', '5786', '5795'],['6000', '5969', '5980', '5989', '5958', '5949'],['8100', '8105', '8110', '8115', '8120', '8125', '8130', '8135', '8140', '8147', '8152', '8157', '8162', '8167', '8172']]
sprite_dict["風莉"]["Ｂ"]["other"] = [['5460', '5112', '5138', '5224', '5294', '5362'],['2292', '2385', '2475', '2504', '2584', '2653', '2726'],['12078', '12083', '12088', '12093', '12098', '12103', '12108', '12113', '12118', '12123', '12135', '12150', '12155', '12165', '12176']]
sprite_dict["美結"]["Ａ"]["other"] = [['2555', '2567', '2579'],['119', '121', '123', '125', '127'],['2917', '2912', '2922', '2927', '2932']]
sprite_dict["七海"]["Ａ"]["other"] = [['866', '900', '932', '967', '1000', '1035'],['2903', '2904', '2905', '2906', '2907', '2908'],['2948', '2953', '2958', '2963', '2968', '2973']]



for c, c_dict in sprite_dict.items():
    for p, p_dict in c_dict.items():
        if len(p_dict["stand"]) == 0:
            continue
        data_dict = {}
        for l in level:
            with open(join(directory, c, c + p + l[:-1] + ".txt"), "r", encoding="utf-16le") as f:
                reader = csv.DictReader(f, delimiter="\t")
                data = list(reader)

            data_dict[l] = {}
            for i in range(len(data)):
                id = 0 if i == 0 else data[i]["layer_id"]
                data_dict[l][id] = data[i]
                data_dict[l][id]["w"] = int(data[i]["width"])
                data_dict[l][id]["h"] = int(data[i]["height"])
                data_dict[l][id]["x"] = int(data[i]["left"]) if i != 0 else 0
                data_dict[l][id]["y"] = int(data[i]["top"]) if i!= 0 else 0

        for stand in p_dict["stand"]:
            elements = [random.choice(sublist) for sublist in p_dict["other"]]
            for l in level:
                base_img = Image.new(mode="RGB", size=(data_dict[l][0]["w"], data_dict[l][0]["h"]), color="gray")
                for e in [stand, *elements]:
                    img = Image.open(join(directory, c, c + p + l + str(e) + ".png"))
                    base_img.paste(img, (data_dict[l][e]["x"], data_dict[l][e]["y"]), img)

                base_img.save(join(directory, l, f"{c}{stand}.png"))
```
