```python
from os.path import join
import os
import misc

directory = "/Users/khoi.ho/Desktop/yorikure_stand/"
directory_out = "/Users/khoi.ho/Desktop/yorikure_stand_result/"

stand_dict = {}
size_list = []
'''
L_kaz_1_d_1ex11.png
  [---v1--][--v2--]
  [---variance----]
'''
for root, dirs, files in os.walk(directory):
    for file in files:
        if file.endswith(".png"):
            size, variance = file.split("_", maxsplit = 1)
            if "_" not in variance:
                continue
            filename = str(join(root, file))
            v1 = variance[:9]
            v2 = variance[9:]
            if size not in size_list:
                size_list.append(size)
            if size != "M":
                continue

            stand_dict.setdefault(v1, {1:[v1+"base.png"], 2:[v1+"base.png"], "face": []})
            if v2 == "base.png":
                pass
            elif "x" not in v2:
                stand_dict[v1]["face"].append(variance)
            elif "ex" not in v2 or ("ex2" in v2 and "mom" in v1):
                stand_dict[v1][int(v2[3])].append(variance)
            else:
                stand_dict[v1][int(v2[3])].insert(0, variance)

# create folder by size list
os.makedirs(directory_out, exist_ok=True)
for size in size_list:
    os.makedirs(join(directory_out, size), exist_ok=True)

#sort face list
for v1 in stand_dict:
    face = stand_dict[v1]['face']
    if len(face) > 0:
        print(v1, len(face))
        stand_dict[v1]['face'] = iter(sorted(face, key = lambda x: os.path.getsize(join(directory, f"S_{x}")), reverse=True))
    else:
        stand_dict[v1]['face'] = None

# add face to list
for v1 in stand_dict:
    face = stand_dict[v1]['face']
    if face == None:
        face = stand_dict[v1[:4]+"1"+v1[5:]]['face']
    for i in [1,2]:
        stand_dict[v1][i].append(next(face))

#blend
for size in size_list:
    for v1 in stand_dict:
        for i in [1,2]:
            # print(f"{v1}_{i}.png", *[f"{size}_{v}" for v in stand_dict[v1][i]])
            img = misc.blend(*[join(directory, f"{size}_{v}") for v in stand_dict[v1][i]])
            img.save(join(directory_out, size, f"{v1}_{i}.png"))

```
